```python
class multifunctions():
    #1.Write a Python program that prints numbers from 1 to 5.
    def Printingnumbers():
        print("Numbers from 1 to 5:")
        for i in range(1,6):
            print(i,end = ' ')
            
    #2.Write a Python program to print the first 5 even numbers.
    def fiveevennumbers():
        print("First 5 even numbers:")
        for numbers2 in range(2,11,2):
            print(numbers2, end = ' ')
                
    #3.Write a Python program to print the sum of numbers from 1 to 10. 
    def sumofnumbers(): 
        number = int(input("Enter the number:"))
        sum = 0
        for item in range(1,number+1):
            sum = sum +item
        print("Sum of given numbers:",sum)
        return sum
        
    #4.Write a Python function to find the factorial of 5.
    def factorial():
        num1 = int(input("Enter the number:"))
        fact = 1
        for i in range(1,num1+1):
            fact = i*fact
        print("Factorial of given number:",fact)   
        
    #5.Write a Python program to check if a number is prime.
    def primenumber(primeno):    
        if primeno <= 1:
            return False
        for i in range(2, int(primeno**0.5) + 1):
            if primeno % i == 0:
                return False
        return True
```
