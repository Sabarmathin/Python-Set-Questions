```python
 #3.Find Even and Odd Numbers in a List
#Question: Write a Python program to separate even and odd numbers from the list [12, 17, 19, 24, 35, 40, 56].
numbers = [12, 17, 19, 24, 35, 40, 56]
evenlist = []
oddlist = []
for item in numbers:
    if item % 2 == 0:
        evenlist.append(item)
    else:
        oddlist.append(item)
print("Evenlist:",evenlist )
print("Oddlist:",oddlist )
```

```python
#4. Reverse a Number
#Question: Write a Python program to reverse the number 12345.
num = 12345
str1 = str(num)
list1 = str(str1)
reverselist = list1[::-1]
print(int(reverselist))
```
